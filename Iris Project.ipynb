# Visualisation

```python
sns.pairplot(iris, hue = 'species' , height = 3 , aspect = 3 );
plt.show()
```

```python
iris.hist(edgecolor='black', linewidth=1.5, figsize=(12,8));
plt.show();
```

```python
plt.figure(figsize=(12,8));
plt.subplot(2,2,1)
sns.violinplot(x='species', y='sepal_length',facecolor = 'red' ,data=iris)
plt.subplot(2,2,2)
sns.violinplot(x='species', y='sepal_width',facecolor = 'yellow' , data=iris)
plt.subplot(2,2,3)
sns.violinplot(x='species', y='petal_length',facecolor = 'green' , data=iris)
plt.subplot(2,2,4)
sns.violinplot(x='species', y='petal_width',facecolor = 'indigo' , data=iris);
plt.show()
```

```python
iris.boxplot(by='species', figsize=(12,8));
plt.show()
```

```python
pd.plotting.scatter_matrix(iris, figsize=(12,10))
plt.show()
```

```python
sns.pairplot(iris, hue="species", diag_kind="kde" , height = 3 , aspect = 1);
plt.show()
```
